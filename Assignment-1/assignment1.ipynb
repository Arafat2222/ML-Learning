# Visualization

```python
plt.rcParams.update({'font.size':22})
#plt.rc('xtick',labelsize = 100)

plt.figure(figsize = (15,12))
plt.scatter(df.Height, df.Weight)
plt.xlabel("Height",fontsize = 30,color = "blue")
plt.ylabel("Weight",fontsize = 30,color = "blue")
plt.title("Height Vs Weight", fontsize = 30)
```

```python
# x axes =feature -> write [[" x "]]/[[' x ']]
# y axes = predict -> [" y "]/[' y']
```

```python
x = df[['Height']]
y = df ['Weight']
```

```python
#print("X\n",x.head(),"\nY\n",y.head())
print("X\n",x.head(),"\nY\n",y.head())
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain,xtest,ytrain,ytest = tts(x,y,test_size = 0.3,random_state = 5)
```

```python
print(xtrain.head())
```

```python
print(ytrain.head())
```

```python
print(xtest.head())
```

```python
print(ytest.head())
```

```python
#test row
trow,tcol = xtest.shape
print("Test Row :",trow)
```

```python
#train row
trrow,ttcol = xtrain.shape
print("Train Row :",trrow)
```

# Linear Regrassion

```python
from sklearn.linear_model import LinearRegression
main_obj = LinearRegression()
```

```python
main_obj.fit(xtrain,ytrain)
```

```python
# coefficient
m = main_obj.coef_ 
print(m)
```

```python
#constant 
c = main_obj.intercept_
print(c)
```

```python
#regression val(pred), using formula
y = m*73.66527364 + c
print(y)
```

```python
#regression val(pred), using function
pred_w = main_obj.predict([[73.66527364]])
print(pred_w)
```

```python
pred_w = main_obj.predict(xtest)
print(pred_w)
```

```python
df["Predicted Weight"] = main_obj.predict(x)
df.head()
```

# Calculate Error 

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mse = mean_squared_error(ytest,pred_w)
print(mse)
```

```python
mae = mean_absolute_error(ytest,pred_w)
print(mae)
```

# Best Fit Line
